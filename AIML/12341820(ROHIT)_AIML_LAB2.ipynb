```python
import noisereduce as nr
from scipy.io import wavfile
# 1. Load the noisy audio file and its sample rate
rate, data = wavfile.read("my_noisy_audio.wav")
# 2. Select a sample of pure noise
noise_clip = data[0:rate]
# 3. Perform noise reduction
reduced_noise_data = nr.reduce_noise(y=data, sr=rate, y_noise=noise_clip)
# 4. Save the clean audio to a new file
wavfile.write("my_cleaned_audio.wav", rate,
reduced_noise_data)
print("Noise reduction complete!")
```

```python
import os

audio_files_to_transcribe = [
    "my_clean_audio (2).wav",
    "noisyfinal.wav",
    "my_cleaned_audio.wav"
]

for audio_file in audio_files_to_transcribe:
    if os.path.exists(audio_file):
        print(f"Transcribing {audio_file}...")
        result = model.transcribe(audio_file)
        print(f"Transcription of {audio_file}:\n", result["text"])
    else:
        print(f"File not found: {audio_file}")
```

```python
import jiwer

# Define the ground truth transcription (original sentence)
ground_truth = "The quick brown fox jumps over the lazy dog."

# Get the noisy and cleaned transcriptions from the previous output
noisy_transcription = "The quick brown box jumps over the lazy socks."
cleaned_transcription = "The quick brown fox jumps over the lazy dog."

# Calculate WER for the noisy transcription
wer_noisy = jiwer.wer(ground_truth, noisy_transcription)

# Calculate WER for the cleaned transcription
wer_cleaned = jiwer.wer(ground_truth, cleaned_transcription)

# Print the WER percentages
print(f"Word Error Rate (WER) for noisy transcription: {wer_noisy:.2%}")
print(f"Word Error Rate (WER) for cleaned transcription: {wer_cleaned:.2%}")
```
